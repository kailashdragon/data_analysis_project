## Exploratory Data Analysis

### RQ 1: Which year has the highest releases of movies

```python
import seaborn as sns
movies_by_year = tmdbdf.groupby('release_year').count()['id']
print(movies_by_year.tail())

movies_by_year.plot(xticks=np.arange(1960, 2015, 5))

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Years vs number of movies', fontsize=14)
plt.xlabel('Release years', fontsize=12)
plt.ylabel('Number of movies', fontsize=12)
sns.set_style('whitegrid')
```

### RQ 2: which movie has the highes and lowest profit

```python
tmdbdf['profit'] = tmdbdf['revenue']-tmdbdf['budget']

def findminmax(x):
    low_index = tmdbdf[x].idxmin()
    high_index = tmdbdf[x].idxmax()
    low = pd.DataFrame(tmdbdf.loc[low_index,:])
    high = pd.DataFrame(tmdbdf.loc[high_index,:])
    
    print('Movie which has low '+x+" : ", tmdbdf['original_title'][low_index])
    print('Movie which has high '+x+" : ", tmdbdf['original_title'][high_index])
    
    return pd.concat([low, high], axis=1)

findminmax('profit')
```

```python
# top 10 popular movies
top_ten = pd.DataFrame(tmdbdf['profit'].sort_values(ascending=False))
top_ten['original_title']=tmdbdf['original_title']
data = list(map(str, (top_ten['original_title'])))
x = list(data[:10])
y = list(top_ten['profit'][:10])

sns.pointplot(x=y, y=x)
```
