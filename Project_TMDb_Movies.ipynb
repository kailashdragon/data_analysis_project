## Exploratory Data Analysis

### RQ 1: Which year has the highest releases of movies

```python
import seaborn as sns
movies_by_year = tmdbdf.groupby('release_year').count()['id']
print(movies_by_year.tail())

movies_by_year.plot(xticks=np.arange(1960, 2015, 5))

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Years vs number of movies', fontsize=14)
plt.xlabel('Release years', fontsize=12)
plt.ylabel('Number of movies', fontsize=12)
sns.set_style('white')
```

### RQ 2: which movie has the highes and lowest profit

```python
tmdbdf['profit'] = tmdbdf['revenue']-tmdbdf['budget']

def findminmax(x):
    low_index = tmdbdf[x].idxmin()
    high_index = tmdbdf[x].idxmax()
    low = pd.DataFrame(tmdbdf.loc[low_index,:])
    high = pd.DataFrame(tmdbdf.loc[high_index,:])
    
    print('Movie which has low '+x+" : ", tmdbdf['original_title'][low_index])
    print('Movie which has high '+x+" : ", tmdbdf['original_title'][high_index])
    
    return pd.concat([low, high], axis=1)

findminmax('profit')
```

```python
# top 10 popular movies
top_ten = pd.DataFrame(tmdbdf['profit'].sort_values(ascending=False))
top_ten['original_title']=tmdbdf['original_title']
data = list(map(str, (top_ten['original_title'])))
x = list(data[:10])
y = list(top_ten['profit'][:10])

sns.pointplot(x=y, y=x)

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Top 10 Popular Movies', fontsize=14)
plt.xlabel('Profit of the movies for the year', fontsize=12)
plt.ylabel('Movie name', fontsize=12)
sns.set_style('dark')
```

### RQ 3: Which year has the highest profit rate

### RQ 10: Which length movies most liked by the audiences according to their popularity?

### RQ 12: How Does The Revenue And Popularity differs Budget And Runtime? And How Does Popularity Depends On Profit?

```python
ax = sns.regplot(x = tmdbdf['revenue'], y = tmdbdf['budget'], color='c')

sns.set(rc={'figure.figsize':(20,5)})
ax.set_title('revenue vs budget graph', fontsize=14)
ax.set_xlabel('revenue', fontsize=12)
ax.set_ylabel('budget', fontsize=12)
sns.set_style('darkgrid')

tmdbdf['revenue'] = tmdbdf['revenue'].replace(0, np.nan)
tmdbdf['budget'] = tmdbdf['budget'].replace(0, np.nan)
tmdbdf['profit'] = tmdbdf['profit'].replace(0, np.nan)

correlated_data = tmdbdf.corr()
print('Correlation between revenue and budget is: ', correlated_data.loc['revenue', 'budget'])
```

```python
ax = sns.regplot(x = tmdbdf['release_year'], y = tmdbdf['vote_average'], color='c')

sns.set(rc={'figure.figsize':(6,4)})
ax.set_title('Release year vs Rating graph', fontsize=14)
ax.set_xlabel('Release Year', fontsize=12)
ax.set_ylabel('Rating', fontsize=12)
sns.set_style('darkgrid')

correlated_data = tmdbdf.corr()
print('Correlation between Release year and Rating is: ', correlated_data.loc['release_year', 'vote_average'])
```

```python
ax = sns.regplot(x = tmdbdf['popularity'], y = tmdbdf['profit'], color='c')

sns.set(rc={'figure.figsize':(6,4)})
ax.set_title('Popularity vs Profit graph', fontsize=14)
ax.set_xlabel('Popularity', fontsize=12)
ax.set_ylabel('Profit', fontsize=12)
sns.set_style('darkgrid')

correlated_data = tmdbdf.corr()
print('Correlation between popularity vs Profit is: ', correlated_data.loc['popularity', 'profit'])
```

#### From this, we can conclude that:
1. There is positive correlation between budget and revenue which means movies which has spent huge busget has greater revenue as compared to the movies which spent low budget.

2. There is negative correlation between the release year and average rating by people. Not all movies which released each year will have good rating. Sometimes increase, sometimes decrease.

3. There is positive correlation between popularity of movie and profit. Movies which has good popularity are likely to gain good profit.

4. If we take the first and third conclusion, we can say that movies who spent more budget are likely to gain more profit.

### RQ13: Which genre has highest release of movies 

```python
def genre_count(x):
    data = tmdbdf[x].str.cat(sep='|')
    glist = pd.Series(data.split('|'))
    info = glist.value_counts(ascending=True)
    
    return info

genre_total = genre_count('genres')
genre_total.plot(kind='bar', figsize=(12,6), fontsize=12, colormap='Set1')

plt.title("Genre With Highest Release",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=13)
plt.ylabel("Genres",fontsize= 13)
sns.set_style("darkgrid")
```

```python
i = 0
genre_count = []
for genre in genre_total.index:
    genre_count.append([genre, genre_total[i]])
    i = i+1
    
plt.rc('font', weight='bold')
f, ax = plt.subplots(figsize=(5, 5))
genre_count.sort(key = lambda x:x[1], reverse = True)
labels, sizes = zip(*genre_count)
labels_selected = [n if v > sum(sizes) * 0.01 else '' for n, v in genre_count]
ax.pie(sizes, labels=labels_selected,
       autopct = lambda x:'{:2.0f}%'.format(x) if x > 1 else '',
       shadow=False, startangle=0)
ax.axis('equal')
plt.tight_layout()
```

From this, we can say that highest release is for the **drama** type of movies followed by **comedy and thriller**.
